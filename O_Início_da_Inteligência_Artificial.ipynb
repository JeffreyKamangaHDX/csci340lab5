### 1. 1943: McCulloch-Pitts (M-P) – A Lógica do Neurônio**

  Em 1943, o neurofisiologista Warren McCulloch e o lógico Walter Pitts
publicaram o artigo seminal ["A Logical Calculus of the Ideas Immanent in Nervous Activity"](chrome-extension://efaidnbmnnnibpcajpcglclefindmkaj/https://www.cs.cmu.edu/~epxing/Class/10715/reading/McCulloch.and.Pitts.pdf). Eles buscavam responder à pergunta: **Como o cérebro pode ser entendido como um dispositivo de computação?**

O Teorema deles, essencialmente afirma que qualquer cálculo pode ser realizado por uma rede de neurônios binários simples. McCulloch e Pitts primeiro criaram um modelo radicalmente simplificado de como um neurônio biológico funciona:

    Tudo ou Nada: O neurônio só tem duas opções de estado:
     - Ativado (1) ou;
     - Inativo (0). Não há "meio-termo".

Decisão Simples:  
O neurônio decide se deve se ativar comparando a **soma de suas entradas multiplicadas pelo peso a um limiar (threshold)**. Se a soma das variaveis recebidas for igual ou superior ao limiar, ele dispara. Eles demonstraram que, mesmo sendo tão simples, um único neurônio M-P pode executar as funções lógicas básicas (AND, OR, NOT).  
  
neuronio_mp.png

A saída binária ($Y$) de um neurônio M-P é determinada pela soma de suas entradas excitadoras e inibidoras, em relação a um limiar ($\theta$).$$Y = \begin{cases} 1 & \text{se } \sum_{i} x_i w_i \ge \theta \quad \text{E não houver inibição absoluta} \\ 0 & \text{caso contrário} \end{cases}
$$Onde:
  - $Y$: A saída binária do neurônio (1 para "disparo/ativo", 0 para "inativo").
  - $x_i$: A $i$-ésima entrada para o neurônio (um valor binário: 1 ou 0).
  - $w_i$: O peso sináptico da $i$-ésima entrada. No modelo M-P original, esses pesos são geralmente fixos (ex: 1 para excitação, -1 para inibição).
  - $\sum_{i} x_i w_i$: A soma ponderada das entradas.
  - $\theta$ (theta): O limiar de ativação (ou threshold). É um valor fixo que a soma ponderada deve atingir ou exceder para que o neurônio dispare.

**Exemplo (Porta AND)**

Para um neurônio M-P configurado como uma porta AND com duas entradas ($x_1, x_2$):
  - Pesos: $w_1 = 1, w_2 = 1$  
  
  - $Limiar:$  $\theta$ = 2

O neurônio dispara ($Y=1$) somente se a soma ponderada for 2 ou mais:$$Y = 1 \quad \text{se } (1 \cdot x_1 + 1 \cdot x_2) \ge 2$$Isso só acontece quando $x_1=1$ e $x_2=1$.

### Exemplos em Python Neurônio M-P


```python
# Neurônio McCulloch-Pitts (M-P)
# Limiar (Threshold) fixo
# Pesos fixos em 1

def neuronio_mp(x1, x2, limiar):
    """
    Simula um neurônio McCulloch-Pitts implementando a lógica AND e OR.
    Entradas: x1, x2 (0 ou 1)
    """
    peso_1 = 1
    peso_2 = 1

    # Soma Ponderada
    soma = (x1 * peso_1) + (x2 * peso_2)

    # Função de Ativação (Degrau Binário)
    if soma >= limiar:
        return 1  # Ativado
    else:
        return 0  # Inativado

# Testes
print(f"AND(0, 0): {neuronio_mp(0, 0, 2)}")
print(f"AND(1, 0): {neuronio_mp(1, 0, 2)}")
print(f"AND(1, 1): {neuronio_mp(1, 1, 2)}")
```

```python
print(f"OR(0, 0): {neuronio_mp(0, 0, 1)}")
print(f"OR(1, 0): {neuronio_mp(1, 0, 1)}")
print(f"OR(1, 1): {neuronio_mp(1, 1, 1)}")
```
